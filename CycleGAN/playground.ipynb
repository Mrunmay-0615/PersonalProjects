```python
import PIL
import os
import matplotlib.pyplot as plt
import torch
import torchvision
```

```python
from PIL import Image
```

```python
image = Image.open('../../GTACitySeg/Dataset3/Source_GTA5/images/00003.png')
label = Image.open('../../GTACitySeg/Dataset3/Source_GTA5/labels/00003.png')
tr = torchvision.transforms.Resize(size=(256, 428))
image, label = tr(image), tr(label)
plt.figure()
plt.imshow(image)
plt.figure()
plt.imshow(label)
```

```python
image = Image.open('../../GTACitySeg/Dataset3/Target_CityScapes/leftImg8bit/train/bochum/bochum_000000_000313_leftImg8bit.png')
label = Image.open('../../GTACitySeg/Dataset3/Target_CityScapes/gtFine/train/bochum/bochum_000000_000313_gtFine_color.png')
tr = torchvision.transforms.Resize(size=(256, 428))
image, label = tr(image), tr(label)
plt.figure()
plt.imshow(image)
plt.figure()
plt.imshow(label)
```

```python
sizes = []
for fp in os.listdir('../../GTACitySeg/Dataset3/Target_CityScapes/leftImg8bit/train/bochum/'):
    if 'DS_Store' in fp:
        continue
    fp = os.path.join('../../GTACitySeg/Dataset3/Target_CityScapes/leftImg8bit/train/bochum/', fp)
    image = Image.open(fp)
    if image.size not in sizes:
        sizes.append(image.size)

print(sizes)
```
